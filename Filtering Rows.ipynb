#### How to filter rows of pd.df by column value

```python
import pandas as pd
```

```python
movies=pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
# Method 1: 

movies[movies.duration < 200]
```

```python
# Method 2: can do the same by passing a list

is_long = movies.duration < 200
movies[is_long]
```

```python
# Method 3: creating a boolean list for duration atleast 200

#creating a empty list 
boolean=[]
for l in movies.duration:
    if l>=200:
        boolean.append(True)
    else:
        boolean.append(False)        
```

```python
#checking first 5 rows of a list
boolean[0:5]
```

```python
# converting a list to series
length=pd.Series(boolean)
```

```python
# Passing the series of boolean values to the dataframe, which in turn returns only the rows with True value.

movies[length]
```

```python
# Filtering the rows above 9 rating
movies[movies.star_rating > 9]
```

```python
# Getting the genre column by filtering the rows with more star_rating
movies[movies.star_rating > 9].genre

# or movies[movies.star_rating > 9][genre]
```

```python
# But using loc[] is the best way

movies.loc[movies.star_rating >= 9, ['title','genre']]
```

#### How to apply Multiple filter criteria to a pd.df

```python
movies.head()
```

```python
movies[(movies.star_rating >=7) & (movies.genre =='Drama')]
```

```python
movies[(movies.star_rating >= 9) | (movies.content_rating =='PG-13')]
```

```python
# Using .isin() to filter a list of strings in a column

movies[(movies.genre.isin(['Action','Drama'])) &(movies.star_rating >=8.5)]
```
